```python
import sympy as sp

# Создаем матрицы
A = sp.MatrixSymbol('A', 3, 3)
B = sp.MatrixSymbol('B', 3, 3)
I = sp.Identity(3)  # Единичная матрица 3x3

# Формула для матрицы входов-выходов A
A_formula = sp.Eq(A, I - B)

# Формула для матрицы коэффициентов прямых и косвенных затрат B
B_formula = sp.Eq(B, I - A)

# Вывод формул
print("Формула для матрицы входов-выходов A:")
sp.pprint(A_formula, use_unicode=True)
print("\nФормула для матрицы коэффициентов прямых и косвенных затрат B:")
sp.pprint(B_formula, use_unicode=True)
```
