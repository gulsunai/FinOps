```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Создание DataFrame на основе вымышленных данных
data = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-05', '2023-01-10', '2023-01-15', '2023-01-20',
             '2023-01-25', '2023-02-01', '2023-02-05', '2023-02-10', '2023-02-15',
             '2023-02-20', '2023-02-25', '2023-03-01', '2023-03-05', '2023-03-10',
             '2023-03-15', '2023-03-20'],
    'Expenses': [100.50, 25.00, 200.00, 50.75, 150.00, 75.20, 120.00, 30.00, 60.50, 180.00, 45.25, 110.00,
                 40.00, 75.90, 210.00, 60.25, 130.00],
    'Category': ['Groceries', 'Dining Out', 'Shopping', 'Utilities', 'Entertainment', 'Groceries',
                 'Dining Out', 'Shopping', 'Utilities', 'Entertainment', 'Groceries', 'Dining Out',
                 'Shopping', 'Utilities', 'Entertainment', 'Groceries', 'Dining Out'],
    'Payment Method': ['Credit Card', 'Cash', 'Debit Card', 'Bank Transfer', 'Credit Card', 'Cash',
                       'Credit Card', 'Debit Card', 'Bank Transfer', 'Credit Card', 'Cash', 'Credit Card',
                       'Debit Card', 'Bank Transfer', 'Credit Card', 'Cash', 'Credit Card']
})

# Шаг 2: Изучение данных
# Выведите первые несколько строк, чтобы понять структуру данных:
print(data.head())

# Шаг 3: Общая информация о данных
# Определите общую информацию о данных, такую как типы столбцов и наличие пропущенных значений:
print(data.info())

# Шаг 4: Визуализация данных
# Создайте график временного ряда затрат
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Expenses', data=data)
plt.title('Expenses Over Time')
plt.xlabel('Date')
plt.ylabel('Expenses')
plt.xticks(rotation=45)
plt.show()

# Шаг 5: Создание круговой диаграммы категорий расходов
plt.figure(figsize=(8, 8))
category_expenses = data.groupby('Category')['Expenses'].sum()
plt.pie(category_expenses, labels=category_expenses.index, autopct='%1.1f%%', startangle=90)
plt.title('Expense Distribution by Category')
plt.axis('equal')  # Равные осям для создания круговой диаграммы
plt.show()

# Шаг 6: Гистограмма расходов
plt.figure(figsize=(10, 6))
sns.histplot(data['Expenses'], kde=True, bins=10)
plt.title('Expense Distribution Histogram')
plt.xlabel('Expenses')
plt.ylabel('Frequency')
plt.show()

# Шаг 7: Анализ данных
# Выполните анализ данных, например, вычисление среднего расхода:
average_expenses = data['Expenses'].mean()
print(f'Average Expenses: ${average_expenses:.2f}')
```
